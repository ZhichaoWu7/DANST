## 4. Visualization

```python
import pandas as pd
import numpy as np
predict_result = pd.read_table("/home/featurize/work/DANST_model/DANST/result/MBC/target_predicted_fraction.csv", sep=',', header=0, index_col=0,encoding="utf-8")

def extract_top_value(predict_result, retain_percent = 0.1): 
    #retain top K% values for each spot
    top_k  = retain_percent * predict_result.shape[1]
    output = predict_result * (np.argsort(np.argsort(predict_result)) >= predict_result.shape[1] - top_k)
    return output

predict_result = extract_top_value(predict_result,retain_percent = 0.5)
cell_types = sc_adata.obs['annotation_1'].unique() 
for i in cell_types:
    ST_adata.obs[i] = predict_result[i].values
```

(1)visualization of spatial distribution of celltypes in st data

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
with mpl.rc_context({'axes.facecolor':  'black',
                     'figure.figsize': [4.5, 5]}):

    sc.pl.spatial(ST_adata,img_key="hires",
                  color=['Ext_L23','Ext_L56','Ext_L5_1',
                       'Ext_L6B','Ext_Hpc_CA1','Ext_Hpc_CA3'],
                  palette='Set1',size=1.5,legend_loc=None)
```

(2) visiualize GT

```python
sc.pl.spatial(ST_adata,
              img_key="hires",
              color=["cluster_rename"],
              title=["cluster_rename"],
              show=True)
```

```python
import matplotlib.pyplot as plt
import scanpy as sc
celltypes_to_plot = ['Ext_L25', 'Ext_L23', 'Ext_L56', 'Ext_Hpc_CA1', 
                 'Ext_Hpc_CA3', 'Ext_Hpc_DG1', 'Ext_Thal_1', 'Ext_Pir']

fig, axs = plt.subplots(2, 4, figsize=(24, 12))
axs = axs.flatten()


for i, celltype in enumerate(celltypes_to_plot):
    sc.pl.violin(ST_adata, celltype, groupby='cluster_rename', 
                  stripplot=False, inner='box', ax=axs[i], show=False, xlabel=None)
    axs[i].tick_params(axis='x', rotation=65, labelsize=18)  
    axs[i].set_ylabel(celltype, fontsize=18) 
    axs[i].set_title(celltype, fontsize=20) 
plt.tight_layout()
plt.show()
```
